## Example Plot

````python
with tp.teed(
    sns.relplot,
    data=df,
    x="horsepower",
    y="mpg",
    hue="origin",
    size="weight",
    alpha=0.5,
    height=6,
    legend=False,
    sizes=(40, 400),
    palette="muted",
    teeplot_outexclude="palette",
    teeplot_show=True,
    teeplot_subdir=pathlib.Path(__file__).stem,
) as g:
    g.figure.set_size_inches(5, 2)
````
